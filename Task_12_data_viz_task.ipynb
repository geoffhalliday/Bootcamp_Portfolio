```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load data
car_df = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile
# Dataframe filtered to specified manufacturers, and stored in 'car_sample'
car_sample = car_df.loc[car_df['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]
# Boxplot produced with title added and colour to brighten appearance
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=car_sample, hue='Manufacturer').set_title('Revs per mile (Audi, Hyundai, Suzuki & Toyota)')
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota


### 2. Histogram

```python
# Histogram of MPG
labels = ('City', 'Highway')
x = car_df['MPG.city'].values
y = car_df['MPG.highway'].values
# Histograms plotted with 50% transparency so both are visable
plt.hist(x,alpha=0.5)
plt.hist(y,alpha=0.5)
# Title and legend displayed to explain the data shown
plt.title("Miles per Gallon (City vs Highway driving)")
plt.legend(labels)

```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway? 

**Histogram answer:**
On the highway

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

# Lineplot drawn comparing 'Wheelbase' and 'Turnn.circle'
lp = sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_df)
# Title added
lp.set_title('Wheelbase vs Turning Circle Size')
# Y-axis label set to 'Turning Circle' instead of the default display of 'Turn.circle'
lp.set_ylabel('Turning Circle')

```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
There is a positive correlation between wheelbase and turning circle.  As the wheelbase gets larger the turning circle gets larger.


### 4. Barplot

```python
# Barplot for mean horsepower
# Car types listed in order of size (although I'm not sure where 'Compact' and 'Sporty' should be positioned) and 'hue' used to add colour
# I could have omitted 'Van' as this is not a car type, but as the dataset is 'Cars93' not 'Vehicles93', I have taken the term car to include vans
bp = sns.barplot(x='Type', y='Horsepower', data=car_df, order=['Small','Compact','Sporty','Midsize','Large','Van'], hue='Type')
# Title added
bp.set_title('Type of Car vs Horsepower')
# X-axis label changed to 'Size of Car'
bp.set_xlabel('Size of Car')
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
With the exception of vans, barplot shows that larger cars on average have higher horsepower.  So, as a general rule, yes a larger car does have more horsepower, with the exception of vans (if these are to be considered and regarded larger than a large car), and also assuming i have positioned 'Compact' and 'Sporty' correctly in the order of size.
